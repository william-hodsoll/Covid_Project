```python
# Import pandas for data manipulation

import pandas as pd
```

```python
# set up options for better viewing of the data

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# read in the data using PyArrow as the dtype backend: PyArrow has better integration and functionality

data = pd.read_csv('/home/william/Desktop/covid_raw.csv', dtype_backend='pyarrow')
```

```python
# drop columns I'm not going to be using

mod_data = data.drop(columns=['new_cases_smoothed', 'new_deaths_smoothed', 'new_deaths_smoothed_per_million', 'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
                              'new_vaccinations_smoothed', 'new_vaccinations_smoothed_per_million', 'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred'])
```

```python
# "tests units' is a string data type and needs to be converetd so I can fill in the nulls with 0's later

mod_data1 = pd.get_dummies(data=mod_data, columns=['tests_units'], dtype=int)
```

```python
# the first 4 columns are strings/dates so I want to exclude them from the next step.
# this cell is just confirming that.

nonfill_data = mod_data.iloc[:, :4]
nonfill_data.head()
```

```python
# filling in nulls with 0's
fill_data = mod_data.iloc[:, 4:].drop(columns=['tests_units']).fillna(0)
```

```python
# rejoining the data

final_data = nonfill_data.join(fill_data)
```

```python
final_data.head()
```

```python
# reports were gathered every 7 days, but data listed (example: total_deaths) as a running sum. This caused a problem later in SQL when trying to sum the total_deaths as it would
# try to sum everyday and give us inflated numbers. This code takes every 7th row, so we avoid that problem later.

final_data = final_data.groupby(['total_deaths']).nth(0).reset_index(drop=True)

# final_data.to_csv('/home/william/Desktop/SQL_project/python_scripts/final_data_complete.csv')

final_data.head()
```

```python
# the following code blocks are seperating out countries from continents and some income levels and gets there max about of deaths
#  saves this info in a dataframe for easy visualization later

location_list = list(final_data['location'].values.unique())
total_deaths_list = []
```

```python
for location in location_list:
    total_deaths_list.append(final_data[final_data['location'] == location]['total_deaths'].max())
```

```python
total_deaths_df = pd.DataFrame(list(zip(location_list, total_deaths_list)), columns=['location', 'total_deaths']) 

total_deaths_df.head()
```

```python
continent_locations = ['Africa', 'Europe', 'Asia', 'European Union', 'High income', 'Low income', 'Lower middle income', 'North America', 'South America', 'World', 'Upper middle income', 'Oceania']
```

```python

total_deaths_df_countries = total_deaths_df[~total_deaths_df['location'].isin(continent_locations)]
```

```python

# Total Deaths by Country to a csv
total_deaths_df_countries.to_csv('/home/william/Desktop/SQL_project/python_scripts/total_deaths_df_by_country.csv')
```
